# 🧠 Simple LLM Text Generation with DistilGPT2

This notebook shows how to use a small language model (LLM) to generate text given a prompt using Hugging Face Transformers.

- Model: `distilgpt2` (a light version of GPT-2)
- Task: Text Generation
- Tools: Python, Hugging Face 🤗, Transformers

```python
# 📦 Install Transformers (if needed)
!pip install transformers
```

```python
# 🔧 Load the text-generation pipeline
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
```

```python
# ✏️ Input your prompt here
prompt = "The future of AI is"

# 🧠 Generate text
results = generator(prompt, max_length=50, num_return_sequences=1)
print(results[0]['generated_text'])
```
